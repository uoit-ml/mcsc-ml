# Kmeans exercise using scipy

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons, make_circles, make_classification, make_blobs
```

```python
from matplotlib.colors import ListedColormap
cm = plt.cm.RdBu
cm_bright = ListedColormap(['#FF0000', '#0000FF'])
cm_bright_5 = ListedColormap(['#FF0000', '#0000FF', '#FFF000', '#000FFF', '#FF00FF'])
```

# Make classification

```python
X, y = make_classification(n_features=2, n_redundant=0, n_informative=2, random_state=1, n_clusters_per_class=1)
plt.figure(figsize=(7,7))
plt.scatter(X[:,0], X[:,1], c=y, cmap=cm_bright)
```

# Make Circles

```python
X, y = make_circles(noise=0.1, factor=.2, random_state=1)
plt.figure(figsize=(7,7))
plt.scatter(X[:,0], X[:,1], c=y, cmap=cm_bright)
```

# Make moons

```python
X, y = make_moons(noise=0.1, random_state=3)
plt.figure(figsize=(7,7))
plt.scatter(X[:,0], X[:,1], c=y, cmap=cm_bright)
```

# Make Blobs

```python
n_samples = 1500
random_state = 170
n_centers = 5
X, y = make_blobs(n_samples=n_samples, centers=n_centers, random_state=random_state)
plt.figure(figsize=(7,7))
plt.scatter(X[:,0], X[:,1], c=y, cmap=cm_bright_5)
```

# Clustering

## Meanshift

```python
from sklearn.cluster import MeanShift
meanshift = MeanShift(bandwidth=2).fit(X) # Try bandwidth 0.1, 1 and 2
plt.figure(figsize=(7,7))
plt.scatter(X[:,0], X[:,1], c=meanshift.labels_, cmap=cm)
```

## KMeans

# "make circle" data

```python
X, y = make_circles(noise=0.1, factor=.2, random_state=1)
plt.figure(figsize=(7,7))
plt.scatter(X[:,0], X[:,1], c=y, cmap=cm_bright)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2, random_state=0).fit(X) # Try bandwidth 0.1, 1 and 2
plt.figure(figsize=(7,7))
plt.scatter(X[:,0], X[:,1], c=kmeans.labels_, cmap=cm)
```

```python
kmeans = KMeans(n_clusters=2, random_state=0).fit(np.hstack([np.sum(X**2, axis=1, keepdims=True)]))
plt.figure(figsize=(7,7))
plt.scatter(X[:,0], X[:,1], c=kmeans.labels_, cmap=cm)
```
