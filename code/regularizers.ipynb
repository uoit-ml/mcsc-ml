# Regularizers

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Vector norm

```python
def vec_p_norm(v,p):
    """
    Returns p-norm of a vector.  See (http://mathworld.wolfram.com/VectorNorm.html) 
    for more details.
    
    Arguments:
        v -- a dimensional vector (if a matrix, treats columns as vectors)
        p -- norm
        
    Returns:
        p-norm of a vector
        
    """
    return np.sum(np.abs(v)**p, axis=0)**(1./p)
```
