# Sklearn linear regression

# Generating data

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
%matplotlib inline

x = np.linspace(5,15,100)
m, c = .5, 2
y = m * x + c
mu, sigma, n = 0, .4, len(y)
y = y + np.random.normal(mu, sigma, n)

plt.title('Noisy y vs. x')
plt.xlabel('x')
plt.ylabel('y')
plt.axis([0,20,0,12])
plt.plot(x,y,'.')

X = np.ones([len(x), 2])
X[:,0] = x
X[:,1] = x**2
Y = np.ones([len(y), 1])
Y[:,0] = y
```

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()
reg.fit (X, Y)

print reg.coef_
print reg.intercept_

y_estimate = reg.intercept_[0] + reg.coef_[0][0]*np.linspace(0,20,100) +  + reg.coef_[0][1]*np.linspace(0,20,100)*np.linspace(0,20,100)

plt.axis([0,20,0,12])
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x,y,'b.')
plt.plot(np.linspace(0,20,100),y_estimate,'r')
```

```python
from sklearn import linear_model
reg = linear_model.RidgeCV(alphas=[0.1, 1.0, 10.0])
reg.fit(X, Y)       

print reg.coef_
print reg.intercept_

y_estimate = reg.intercept_[0] + reg.coef_[0]*np.linspace(0,20,100)

plt.axis([0,20,0,12])
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x,y,'b.')
plt.plot(np.linspace(0,20,100),y_estimate,'r')

reg.alpha_                                      
```
